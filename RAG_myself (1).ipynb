# **Structured Output**

```python
from typing import List
from pydantic import BaseModel, Field

class MobileReview(BaseModel):
    phone_model: str = Field(description="Name and model of the phone")
    rating: float = Field(description="Overall rating out of 5")
    pros: List[str] = Field(description="List of positive aspects")
    cons: List[str] = Field(description="List of negative aspects")
    summary: str = Field(description="Brief summary of the review")

review_text = """
Just got my hands on the new Galaxy S21 and wow, this thing is slick! The screen is gorgeous,
colors pop like crazy. Camera's insane too, especially at night - my Insta game's never been
stronger. Battery life's solid, lasts me all day no problem.

Not gonna lie though, it's pretty pricey. And what's with ditching the charger? C'mon Samsung.
Also, still getting used to the new button layout, keep hitting Bixby by mistake.

Overall, I'd say it's a solid 4 out of 5. Great phone, but a few annoying quirks keep it from
being perfect. If you're due for an upgrade, definitely worth checking out!
"""

structured_llm = llm.with_structured_output(MobileReview)
output = structured_llm.invoke(review_text)
output
```

# **Specific pdf niye kaj korbo sejonno ekta specific pdf file re read kore split korailam 30 ta chunk e vag korlam pura pdf ta**

```python
import os
from langchain_community.document_loaders import PyPDFLoader, Docx2txtLoader
from langchain_text_splitters import RecursiveCharacterTextSplitter
from langchain_core.documents import Document

# Specify the exact file path
specific_file_path = "/content/drive/MyDrive/Rag_dataset/_OfferLetter_Tasnia_24111101.pdf"

# Initialize the text splitter
text_splitter = RecursiveCharacterTextSplitter(
    chunk_size=200,
    chunk_overlap=50,
    length_function=len
)



if specific_file_path.endswith(".docx"):
  docx_loader = Docx2txtLoader(specific_file_path)
  documents = docx_loader.load()
elif specific_file_path.endswith(".pdf"):
  pdf_loader = PyPDFLoader(specific_file_path)
  documents = pdf_loader.load()
elif specific_file_path.endswith(".txt"):
  with open(specific_file_path, "r", encoding="utf-8") as file:
    text = file.read()
    # Wrap the text into a Document
    documents = [Document(page_content=text)]
else:
  raise ValueError("Unsupported file format.")


# Print the number of loaded documents
print(f"Loaded {len(documents)} documents from {specific_file_path}")

# Split the document into chunks
splits = text_splitter.split_documents(documents)

# Print the number of chunks
print(f"Split the document into {len(splits)} chunks.")


print("\nFirst document (documents[0]):")
# print(documents[0])

# print(splits[0])

# print(splits[0].metadata)

print(splits[0].page_content)
```

```python
from langchain.embeddings import OpenAIEmbeddings
embeddings = OpenAIEmbeddings()

# 4. Embedding Documents

document_embeddings = embeddings.embed_documents([split.page_content for split in splits])

print(f"Created embeddings for {len(document_embeddings)} document chunks.")
```

# **Create and persist Chroma vector store**

```python
from langchain_chroma import Chroma

collection_name = "my_collection"
vectorstore = Chroma.from_documents(
    collection_name=collection_name,
    documents=splits,
    embedding=embedding_function,
    persist_directory="./chroma_db"
)
print("Vector store created and persisted to './chroma_db'")
```

```python
# 5. Perform similarity search

query="who is tasnia"
# query = "What is Blood stain Classification?"
search_results = vectorstore.similarity_search(query, k=2)

print(f"\nTop 4 most relevant chunks for the query: '{query}'\n")
for i, result in enumerate(search_results, 1):
    print(f"Result {i}:")
    print(f"Source: {result.metadata.get('source', 'Unknown')}")
    print(f"Content: {result.page_content}")
    print()
```

```python
retriever = vectorstore.as_retriever(search_kwargs={"k": 1})
# retriever.invoke("What is Blood stain Classification?")

query="状態から致しますと、ファスナーに問題が無く開閉する為のスライダー交換でお直し可能かと存じますが、レーン部分に汚れがある為、交換致しましても、また直ぐに開いてしまう可能性がある為、レーンのお磨き必要な状態です。原因ですが、お財布内に詰めすぎで負荷が掛かった状態で開閉を繰り返した事が原因になるかと存じます。開閉される際に、ファスナー両サイドを摘んで、負荷が掛からないように開閉しますと、長持ちするかと存じます。類似品との交換になり持ち手は再利用させていただきます"
retriever.invoke(query)
```

```python
def docs2str(docs):
    return "\n\n".join(doc.page_content for doc in docs)
```

```python
rag_chain = (
    {"context": retriever | docs2str, "question": RunnablePassthrough()} | prompt
)
rag_chain.invoke("What is Blood stain Classification?")
```
